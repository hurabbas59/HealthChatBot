```python
import requests
import pandas as pd
import numpy as np
from retry import retry  

model_id = "sentence-transformers/all-MiniLM-L6-v2"
hf_token = "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"  

api_url = f"https://api-inference.huggingface.co/pipeline/feature-extraction/{model_id}"
headers = {"Authorization": f"Bearer {hf_token}"}

@retry(delay=10, tries=3)
def query(texts):
    response = requests.post(api_url, headers=headers, json={"inputs": texts, "options": {"wait_for_model": True}})
    return response.json()

df = pd.read_csv("emirates.csv")
vector_col = df["speciality"].tolist()

output = query(vector_col)

# Extract the embeddings from the output
embeddings_array = np.array([item["sequence_output"] for item in output])

# Save the embeddings array to a NumPy file
np.save("embeddings_array.npy", embeddings_array)

# Display the embeddings array
print("Embeddings array:")
print(embeddings_array)
```

```python
import requests
import pandas as pd
import numpy as np
from retry import retry  

model_id = "sentence-transformers/all-MiniLM-L6-v2"
hf_token = "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"  

api_url = f"https://api-inference.huggingface.co/pipeline/feature-extraction/{model_id}"
headers = {"Authorization": f"Bearer {hf_token}"}

@retry(delay=10, tries=3)
def query(texts):
    response = requests.post(api_url, headers=headers, json={"inputs": texts, "options": {"wait_for_model": True}})
    return response.json()

df = pd.read_csv("emirates.csv")
vector_col = df["speciality"].tolist()

output = query(vector_col)

embeddings_array = np.array([item.get("sequence_output", []) for item in output])

np.save("embeddings_array.npy", embeddings_array)

print("Embeddings array:")
print(embeddings_array)
```
